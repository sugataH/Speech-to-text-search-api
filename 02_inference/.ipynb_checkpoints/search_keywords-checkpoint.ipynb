```python
with open("../01_data/corpus.txt", "r", encoding="utf-8") as f:
    corpus_lines = [line.strip() for line in f if line.strip()]
print(f"Loaded {len(corpus_lines)} lines from corpus.")
```

```python
import re
import nltk
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download("punkt")
nltk.download("stopwords")

stop_words = set(stopwords.words("english"))

def search_corpus(transcription, corpus_lines, top_n=5):
    tokens = word_tokenize(transcription.lower())
    keywords = [t for t in tokens if t.isalpha() and t not in stop_words]

    if not keywords:
        return []

    matches = []
    for line in corpus_lines:
        line_lower = line.lower()
        match_count = sum(1 for kw in keywords if re.search(rf"\b{re.escape(kw)}\b", line_lower))
        if match_count > 0:
            matches.append((line, match_count))

    matches.sort(key=lambda x: x[1], reverse=True)

    return [line for line, _ in matches[:top_n]]
```

```python
transcription = "learning from speech representation"
results = search_corpus(transcription, corpus_lines)

print("\nSearch Results:")
for i, line in enumerate(results, 1):
    print(f"{i}. {line}")
```
